# **Edge & Corner Detection | Harris Corner Detection**

## **1. Edge Detection - Understanding Intensity Changes**
An **edge** is a region in an image where the pixel intensity **changes suddenly**.

### **Mathematical Approach**
We compute the **gradient** (rate of intensity change) in x and y directions:

$$
\nabla I = \left( \frac{\partial I}{\partial x}, \frac{\partial I}{\partial y} \right)
$$

- If the **gradient magnitude** is **high**, it indicates an **edge**.
- Example: A transition from **black to white** in an image.

---

## **2. What is a Corner?**
A **corner** is a point where **two edges meet**.  
Unlike an edge (which has intensity change in **one direction**), a corner has **intensity variations in both x and y directions**.

### **Key Observations:**
- **Flat region** â†’ No intensity change in any direction.
- **Edge** â†’ High intensity change in one direction.
- **Corner** â†’ High intensity change in both directions.

---

## **3. Shifting a Window to Detect Corners**
A **window** (small square region) is moved **slightly** in different directions.  

We analyze the **intensity change** when the window moves.

### **Three Possible Cases**
1. **Flat Region** â†’ No change in intensity â†’ Not a corner.
2. **Edge** â†’ Large change in **one direction** â†’ Not a corner.
3. **Corner** â†’ Large change in **both directions** â†’ It's a corner.

---

## **4. Mathematically Modeling Corner Detection**
When the window shifts by **(u, v)**, the intensity change is:

$$
I(x+u, y+v) - I(x, y) \approx u I_x + v I_y
$$

where:
- \( I_x \) and \( I_y \) are the **image intensity derivatives** (gradients).
- If the **change is large in all directions**, it's a **corner**.

---

## **5. Eigenvalues & Their Role in Corner Detection**
We define a **structure tensor matrix \( M \)**:

$$
M = \sum w(x,y)
\begin{bmatrix}
I_x^2 & I_x I_y \\
I_x I_y & I_y^2
\end{bmatrix}
$$

where:
- \( w(x, y) \) is a window function (like Gaussian).
- \( I_x \) and \( I_y \) are intensity derivatives.

The **eigenvalues** \( \lambda_1 \) and \( \lambda_2 \) of this matrix determine:
1. **Flat Region** â†’ Both eigenvalues are **small**.
2. **Edge** â†’ One eigenvalue **large**, the other **small**.
3. **Corner** â†’ Both eigenvalues are **large**.

---

## **6. Harris Corner Detection**
Harris Corner Detection uses **eigenvalues** to determine if a point is a corner.

### **Formula:**
$$
R = \det(M) - k \cdot \text{trace}(M)^2
$$

where:
- \( \det(M) = \lambda_1 \lambda_2 \) (product of eigenvalues).
- \( \text{trace}(M) = \lambda_1 + \lambda_2 \) (sum of eigenvalues).
- \( k \) is a constant (typically 0.04 to 0.06).

### **Interpretation of \( R \) Values**
- \( R > 0 \) (large) â†’ **Corner** (both eigenvalues large).
- \( R < 0 \) (negative) â†’ **Edge** (one eigenvalue much larger).
- \( R \approx 0 \) â†’ **Flat region**.

---

## **7. Harris Corner Detection in Python (Steps)**
1. Convert image to **grayscale**.
2. Compute **image gradients** (\( I_x, I_y \)).
3. Compute the **Harris matrix \( M \)**.
4. Compute **R** using:

   $$
   R = \det(M) - k \cdot \text{trace}(M)^2
   $$

5. Mark pixels as **corners** if \( R \) is large.

# **Why Only Two Eigenvalues in Harris Corner Detection?**

## **1. Introduction**
Harris Corner Detection is a method to find **corners** in an image by analyzing **how intensity changes in different directions**. To do this, we compute a **structure tensor (M)** and analyze its **eigenvalues**.

Since we are working with **2D images**, the matrix \( M \) is always a **\(2 \times 2\) symmetric matrix**, which means it has **exactly two eigenvalues**.

---

## **2. Structure Tensor (M)**
We define the **second-order structure tensor** as:

$$
M = \sum_{(x,y)} w(x,y)
\begin{bmatrix}
I_x^2 & I_x I_y \\
I_x I_y & I_y^2
\end{bmatrix}
$$

where:
- \( I_x \) and \( I_y \) are the **gradients** (rate of intensity change in the x and y directions).
- \( w(x,y) \) is a **weighting function** (like a Gaussian).
- The elements of \( M \) are:

  $$
  A = \sum I_x^2, \quad B = \sum I_x I_y, \quad C = \sum I_y^2
  $$

  So, we can rewrite \( M \) as:

  $$
  M =
  \begin{bmatrix}
  A & B \\
  B & C
  \end{bmatrix}
  $$

---

## **3. Why Only Two Eigenvalues?**
The matrix \( M \) is a **\(2 \times 2\) symmetric matrix**, which always has **exactly two eigenvalues** \( \lambda_1 \) and \( \lambda_2 \). 

These eigenvalues **represent how intensity varies in two perpendicular directions**.

The eigenvalues are found using the characteristic equation:

$$
\det(M - \lambda I) = 0
$$

Expanding this determinant:

$$
\begin{vmatrix}
A - \lambda & B \\
B & C - \lambda
\end{vmatrix}
= 0
$$

Solving for \( \lambda \), we get:

$$
\lambda = \frac{(A + C) \pm \sqrt{(A - C)^2 + 4B^2}}{2}
$$

Since \( M \) is a **\(2 \times 2\) matrix**, it has only **two** solutions for \( \lambda \), meaning it has exactly **two eigenvalues**.

---

## **4. What Do the Two Eigenvalues Represent?**
The eigenvalues \( \lambda_1 \) and \( \lambda_2 \) tell us **how intensity varies in different directions**.

| **Eigenvalues** | **Interpretation** |
|--------------|----------------|
| **Both small** (\( \lambda_1, \lambda_2 \approx 0 \)) | **Flat region** (no significant intensity change) |
| **One large, one small** (\( \lambda_1 \gg \lambda_2 \) or vice versa) | **Edge** (intensity changes strongly in one direction) |
| **Both large** (\( \lambda_1, \lambda_2 \) are large & similar) | **Corner** (intensity changes in multiple directions) |

- **Edges**: One eigenvalue dominates, meaning the intensity is changing strongly in only **one** direction.
- **Corners**: Both eigenvalues are large, meaning intensity changes **significantly in multiple directions**.

---

## **5. Why Not More Than Two Eigenvalues?**
We are working with a **2D image**, meaning intensity changes are measured along **only two axes (x and y)**. This means:
- We only need to measure intensity variations in **two perpendicular directions**.
- The structure tensor **M** is a **\(2 \times 2\) matrix**, so it has **only two eigenvalues**.

If we were dealing with **3D images (volumetric data)**, then the structure tensor would be **\(3 \times 3\)**, and we'd have **three eigenvalues**.

# Harris Corner Detection and R

In the Harris Corner Detection algorithm, **R** is a function of the eigenvalues (\(\lambda_1\) and \(\lambda_2\)) of the **structure tensor** or **second-moment matrix**. These eigenvalues represent the intensity changes in the **x** and **y** directions around a point.

### The formula for **R** is:

$$
R = \lambda_1 \lambda_2 - k (\lambda_1 + \lambda_2)^2
$$

Where:
- \(\lambda_1\) and \(\lambda_2\) are the **eigenvalues** of the structure tensor, representing intensity changes in two perpendicular directions.
- \(k\) is a constant (typically between 0.04 and 0.06) used to control the sensitivity of corner detection.

### Interpreting **R**:
- **Large R**: When both eigenvalues (\(\lambda_1\) and \(\lambda_2\)) are large (indicating significant intensity changes in both directions), **R** will be **large**, suggesting a **corner**.
- **R > 0**: If **R** is positive, it typically indicates a **corner** (both eigenvalues are large).
- **R â‰ˆ 0**: If **R** is near zero, it suggests a **flat region** (both eigenvalues are small, indicating little to no intensity change).
- **R < 0**: If **R** is negative, it indicates an **edge** (one eigenvalue is much larger than the other, indicating intensity change in only one direction).

### Summary:
- **R** is used to classify a point based on the eigenvalues of the structure tensor.
- **R** helps identify whether the point is a **corner** (large eigenvalues, large **R**), **edge** (one large eigenvalue, negative **R**), or **flat region** (small eigenvalues, **R** close to zero).
