# Limits, Continuity, and L'H么pital's Rule

## **1. Limits**

A limit describes the behavior of a function as the input approaches a particular value. It is written as:

$$ \lim_{{x \to a}} f(x) = L $$

This means that as \( x \) gets arbitrarily close to \( a \), the function \( f(x) \) approaches the value \( L \).

### **1.1 One-Sided Limits**
- **Right-hand limit:**
  $$ \lim_{{x \to a^+}} f(x) = L $$
  Here, \( x \) approaches \( a \) from values greater than \( a \).
- **Left-hand limit:**
  $$ \lim_{{x \to a^-}} f(x) = L $$
  Here, \( x \) approaches \( a \) from values smaller than \( a \).

For a limit to exist, the left-hand and right-hand limits must be equal:

$$ \lim_{{x \to a^+}} f(x) = \lim_{{x \to a^-}} f(x) = L $$

### **1.2 Special Indeterminate Forms**
When evaluating limits, we may encounter indeterminate forms such as:
- \( \frac{0}{0} \)
- \( \frac{\infty}{\infty} \)
- \( 0 \times \infty \)
- \( \infty - \infty \)

In such cases, algebraic manipulations or L'H么pitals Rule may be required.

---

## **2. Continuity**
A function \( f(x) \) is **continuous at a point** \( x = a \) if:
1. \( f(a) \) is defined.
2. The limit \( \lim_{{x \to a}} f(x) \) exists.
3. \( \lim_{{x \to a}} f(x) = f(a) \)

If any of these conditions fail, \( f(x) \) is **discontinuous** at \( x = a \).

### **2.1 Types of Discontinuities**
1. **Removable Discontinuity**: The function has a hole at \( x = a \), but the limit exists.
2. **Jump Discontinuity**: The left-hand and right-hand limits exist but are not equal.
3. **Infinite Discontinuity**: The function approaches \( \pm \infty \) at \( x = a \).

---

## **3. L'H么pital's Rule**
L'H么pitals Rule helps evaluate limits that result in indeterminate forms like \( \frac{0}{0} \) or \( \frac{\infty}{\infty} \). It states that if:

$$ \lim_{{x \to a}} \frac{f(x)}{g(x)} = \frac{0}{0} \text{ or } \frac{\infty}{\infty} $$

then:

$$ \lim_{{x \to a}} \frac{f(x)}{g(x)} = \lim_{{x \to a}} \frac{f'(x)}{g'(x)} $$

### **3.1 Example: Evaluating \( \lim_{{x \to 0}} \frac{\sin x}{x} \)**
Since directly substituting \( x = 0 \) gives \( \frac{0}{0} \), we differentiate numerator and denominator:

$$ \frac{d}{dx} \sin x = \cos x, \quad \frac{d}{dx} x = 1 $$

Applying L'H么pitals Rule:

$$ \lim_{{x \to 0}} \frac{\sin x}{x} = \lim_{{x \to 0}} \frac{\cos x}{1} = \cos(0) = 1 $$

Thus, \( \lim_{{x \to 0}} \frac{\sin x}{x} = 1 \).
