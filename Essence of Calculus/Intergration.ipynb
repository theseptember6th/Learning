# Integration and the Fundamental Theorem of Calculus

In this discussion, we explore the relationship between integrals and derivatives, focusing on the idea that **integrals are essentially the inverse of derivatives**. We'll also relate this to a real-world problem involving motion and use this example to build intuition about integral problems.

## Introduction to the Problem

Imagine you're sitting in a car, but you can't see outside the window. All you can see is the speedometer. At some point, the car starts moving, speeds up, and then slows down to a stop, all over the course of 8 seconds. The question we want to answer is:  
**Can you figure out how far the car has traveled during this time based only on the speedometer reading?**  
More specifically, can we determine a **distance function** \( s(t) \) that tells us how far the car has traveled after a given amount of time \( t \), where \( t \) is between 0 and 8 seconds?

### Velocity and Distance

Let's assume that the velocity of the car can be described by a function \( v(t) = t(8 - t) \), where \( v(t) \) is in meters per second. We know that in calculus, we can calculate the **velocity** from the **distance function** by taking the derivative:

$$
v(t) = \frac{d}{dt}s(t)
$$

In this case, we want to go the opposite way. We know the **velocity function** \( v(t) \), and we want to figure out the **distance function** \( s(t) \). This means we need to find the **antiderivative** (or integral) of the velocity function.

### General Problem Setup

We begin with a general setup where we need to find the distance traveled by the car, given its velocity function over time. To make it easier to think about, let's assume that the velocity is constant over small intervals. This simplifies the calculation, and we'll later refine this approximation to improve accuracy.

## Visualizing the Problem

Imagine the graph of velocity \( v(t) \) over time. The horizontal axis represents time in seconds, and the vertical axis represents velocity in meters per second.  
The goal is to find the area under the velocity-time graph. This area gives us the total distance traveled by the car over the time interval.

### Constant Velocity Case

If the velocity were constant, say \( v(t) = 7 \) meters per second, finding the distance traveled would be easy:

$$
\text{Distance} = v(t) \times \Delta t
$$

Where \( \Delta t \) is the time interval. The distance traveled can be interpreted as the area of a rectangle under the graph. However, in our case, the velocity is not constant; it changes over time, which complicates the situation.

### Approximating with Rectangles

To handle the varying velocity, we approximate the velocity function as being constant over small intervals. For each small time interval \( \Delta t \), we treat the velocity as constant, taking the velocity value at the start of the interval. The distance traveled during that interval is then:

$$
\text{Distance} = v(t_{\text{start}}) \times \Delta t
$$

This method breaks the total distance into small chunks, each of which is easier to compute because the velocity is constant over the small interval.

### Sum of Areas

To calculate the total distance traveled, we sum up the distances traveled during each small interval. This sum can be written as:

$$
\sum_{i=0}^{n} v(t_i) \times \Delta t
$$

Where \( t_i \) represents the times at which we sample the velocity function, and \( \Delta t \) is the width of each interval. As we make the intervals smaller (i.e., make \( \Delta t \) approach zero), the sum of these areas becomes a more accurate approximation of the total distance traveled.

## Taking the Limit

The key insight is that as \( \Delta t \) gets smaller and smaller, the sum of these rectangular areas approaches a specific value: the **area under the curve** of the velocity function. This area corresponds to the **integral** of the velocity function, which is the total distance traveled.

$$
\text{Total Distance} = \int_0^8 v(t) \, dt
$$

This approach is the heart of integral calculus: breaking up a complicated problem (finding the total distance) into many small, simple pieces (approximating the velocity as constant over small intervals), and then taking the limit as the pieces get infinitesimally small.

## Interpreting the Integral

The integral of the velocity function gives us the distance function. In general, the area under the graph of a function represents the total accumulated quantity (in this case, distance) corresponding to the values of the function (in this case, velocity) over a specified range.

For our specific problem, the integral of \( v(t) = t(8 - t) \) will give us the total distance traveled by the car during the 8-second interval.

### General Rule: Derivative of an Integral

There's a powerful result in calculus known as the **Fundamental Theorem of Calculus**, which connects the process of integration and differentiation. It states that:

$$
\frac{d}{dt} \left( \int_0^t f(u) \, du \right) = f(t)
$$

This tells us that the derivative of the integral of a function \( f(u) \) with respect to \( t \) is simply the original function \( f(t) \). This is why the velocity function is the derivative of the distance function: the **distance function is the integral of the velocity function**, and the **velocity function is the derivative of the distance function**.

## Finding the Antiderivative

Now, let's solve for the distance function by finding the antiderivative of the velocity function:

$$
v(t) = t(8 - t) = 8t - t^2
$$

We need to find a function \( s(t) \) whose derivative is \( 8t - t^2 \). We do this by integrating each term:

- The integral of \( 8t \) with respect to \( t \) is \( 4t^2 \).
- The integral of \( -t^2 \) with respect to \( t \) is \( -\frac{t^3}{3} \).

Therefore, the antiderivative of \( 8t - t^2 \) is:

$$
s(t) = 4t^2 - \frac{t^3}{3} + C
$$

Where \( C \) is a constant of integration. This constant represents the fact that the **distance function** is not uniquely determined without additional information, such as an initial condition (e.g., where the car started from).

## Conclusion

In summary, we used the concept of integrals to determine the total distance traveled by the car based on its velocity function. The process involves approximating the area under the velocity-time curve using rectangles, refining that approximation, and then taking the limit as the rectangle widths become infinitesimally small. This leads us to the integral of the velocity function, which gives us the distance traveled.

The fundamental theorem of calculus helps us connect integration with differentiation, showing that the derivative of the distance function is the velocity function, and vice versa.

### Key Takeaways:

1. **Integration** can be interpreted as finding the total accumulated quantity, such as the total distance traveled.
2. The **area under the curve** of a function can represent a variety of quantities (e.g., distance, area, total work).
3. The **Fundamental Theorem of Calculus** connects **derivatives** and **integrals**.
4. **Antiderivatives** are used to find the distance function when given the velocity function.
