# **Implicit Differentiation & Related Rates**

## **1. The Equation of a Circle**
We start with the equation of a circle centered at the origin with radius 5:

$$
x^2 + y^2 = 5^2
$$

which simplifies to:

$$
x^2 + y^2 = 25
$$

This equation represents all points \((x, y)\) that are exactly 5 units away from the origin.

## **2. Finding the Slope Using Implicit Differentiation**

To find the slope of the tangent line at a given point, we differentiate both sides with respect to \(x\).

### **Step-by-Step Differentiation**
#### **Differentiate Both Sides**
1. Differentiate \(x^2\) using the power rule:
   
   $$
   \frac{d}{dx} (x^2) = 2x
   $$

2. Differentiate \(y^2\) using the chain rule:
   
   $$
   \frac{d}{dx} (y^2) = 2y \frac{dy}{dx}
   $$
   
3. The right-hand side of the equation is a constant (25), so its derivative is 0:
   
   $$
   \frac{d}{dx} (25) = 0
   $$

### **Final Differentiated Equation**

$$
2x + 2y \frac{dy}{dx} = 0
$$

Solving for \(\frac{dy}{dx}\):

$$
\frac{dy}{dx} = \frac{-x}{y}
$$

## **3. Finding the Slope at (3,4)**

Substituting \(x = 3\) and \(y = 4\):

$$
\frac{dy}{dx} = \frac{-3}{4}
$$

So, the slope of the tangent line at \((3,4)\) is **\(-3/4\)**.

## **4. Related Rates Problem (Ladder Problem)**

- A ladder of length **5 meters** leans against a wall.
- Initially:
  - The top of the ladder is **4 meters** high.
  - The bottom is **3 meters** from the wall.
- The top is **slipping down at 1 m/s**.
- We need to find how fast the bottom moves away (\(dx/dt\)).

This forms a **right triangle**:

$$
x^2 + y^2 = 5^2
$$

### **Differentiate Both Sides with Respect to Time**

$$
2x \frac{dx}{dt} + 2y \frac{dy}{dt} = 0
$$

Dividing by 2:

$$
x \frac{dx}{dt} + y \frac{dy}{dt} = 0
$$

Substituting known values \((x = 3, y = 4, dy/dt = -1)\):

$$
3 \frac{dx}{dt} + 4(-1) = 0
$$

$$
3 \frac{dx}{dt} - 4 = 0
$$

$$
3 \frac{dx}{dt} = 4
$$

$$
\frac{dx}{dt} = \frac{4}{3}
$$

So, the bottom of the ladder moves **away from the wall at 4/3 meters per second**.

## **5. Python Code for Calculation**

```python
# Given values
x = 3  # Distance from the wall
y = 4  # Height of the ladder
dy_dt = -1  # Top is moving down at 1 m/s

# Solve for dx/dt
dx_dt = - (y * dy_dt) / x

print(f"The bottom moves away at {dx_dt:.2f} m/s")
```

## **6. Summary**
1. **Implicit differentiation** allows us to differentiate equations where \( y \) is not explicitly written as a function of \( x \).
2. It relies on the **chain rule**, since \( y \) depends on \( x \).
3. In **related rates**, differentiation is done with respect to **time (t)**.
4. Both methods track how changes in one variable affect another.
5. The ladder problem and the circle problem share the same **mathematical structure**.
